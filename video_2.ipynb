```python
from matplotlib import pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
plt.plot(x,y,'D--')# r means red and + means the symbol
```

```python
plt.plot(x,y,color='red',marker='D',linestyle='--',markersize=10)
#D-stands for Diamond
```

```python
plt.plot(x,y,color='red',marker='D',linestyle='--',alpha=0.4)
```
